```python
import pandas as pd  
import numpy as np  
import matplotlib.pyplot as plt  
import seaborn as seabornInstance 
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LinearRegression
from sklearn import metrics
%matplotlib inline
```

```python
data = pd.read_csv(r"C:\Users\AKASH\Desktop\classwork\my_project\Camera_Price_Prediction_dataset.csv")
print("Row, Col:",data.shape)
data.head()
```

```python
data.plot(x='Max resolution',y='Price', style='+')
plt.title('Max resolution VS Price')
plt.xlabel('Max resolution')
plt.ylabel('Price')
print("Data Plotting:",data.plot)
```

```python
x=data['Max resolution'].values.reshape(-1,1)
y=data['Price'].values.reshape(-1,1)
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size= 0.25,random_state= 0)
```

```python
reg= LinearRegression()
reg.fit(x_train, y_train)
```

```python
y_pred=reg.predict(x_test)
```

```python
DF=pd.DataFrame({'Actual':y_test.flatten(),'Prediction':y_pred.flatten()})
DF
```
